Task 3 Product Discount Checker
Requirements

Input:

Product price
Membership status (yes/no)

Rules:

Member + price > 5000 → 20% discount
Non-member + price > 5000 → 10% discount
Otherwise → No discount

```python

price = float(input("Enter product price: "))
membership = input("Are you a member? (yes/no): ").lower()

if membership == "yes" and price > 5000:
    discount = price * 0.20  # 20% discount for members
elif membership == "no" and price > 5000:
    discount = price * 0.10  # 10% discount for non-members
else:
    discount = 0  # No discount
final_price = price - discount
print(f"\nOriginal Price: Rs. {price}")
print(f"Discount Applied: Rs. {discount}")
print(f"Final Price: Rs. {final_price}")
```

Project : AI Recruitment Decision System
Scenario

A software house wants an intelligent hiring system.

Inputs
Name
Age
CGPA
Skills Count
Experience
Expected Salary
Communication Score (1–10)
Logic
Reject Conditions

Reject if:

Age < 18
CGPA < 2.7
Skills < 3
Fraud Detection

Flag as suspicious if:

Expected Salary > 400000
AND Experience = 0
Hiring Levels
Intern
CGPA ≥ 2.7
Skills ≥ 3
Junior Developer
CGPA ≥ 3.0
Skills ≥ 5
Experience ≥ 1
Software Engineer
CGPA ≥ 3.3
Skills ≥ 7
Communication ≥ 7

```python
name = input("enter your name")
age = int(input("enter your age"))
cgpa = float(input("enter your cgpa"))
Skills_Count = int(input("enter your skills count"))
experience = int(input("enter your experince in years"))
Expected_Salary = int (input("enter your expected salary"))
Communication_Score = int(input("enter your communication score out of 10"))

if age < 18:
    print(name, "Rejected (Reason: Age < 18)")

elif cgpa < 2.7:
    print(name, " Rejected (Reason: CGPA < 2.7)")

elif Skills_Count < 3:
    print(name, " Rejected (Reason: Skills count < 3)")

elif Expected_Salary > 400000 and experience == 0:
    print(name, "Flagged as Suspicious (High Salary + No Experience)")

elif cgpa >= 3.3 and Skills_Count >= 7 and Communication_Score >= 7:
    print(name, "Hired as Software Engineer")

elif cgpa >= 3.0 and Skills_Count >= 5 and experience >= 1:
    print(name, " Hired as Junior Developer")

elif cgpa >= 2.7 and Skills_Count >= 3:
    print(name, " Hired as Intern")

else:
    print(name, "Not Selected")


```
