1. You have to develop a program that performs the following tasks:

Create variables to store:
University name
Total number of students
Scholarship budget
Passing marks

2. Create a list containing the names of at least 8 students.

3. Create a tuple that stores available departments:

	("CS", "SE", "AI", "DS", "IT")

4.Use a loop to ask the user to enter:
Student name
Department
Marks obtained
Attendance percentage
The program should continue taking input until the user enters "exit" as the student name.
Use the break statement to stop the loop.

5.Apply the following if conditions:
If marks are greater than or equal to passing marks AND attendance is greater than 75:
Print "Eligible for Exam"
If marks are above 85 and attendance is greater than 80:
Print "Scholarship Candidate"
If attendance is below 50 or marks less than 50:
Print "Student Not eligible"
If department entered is not present in the tuple:
Print "Invalid Department"

6. Store all eligible students in a new list called passed_students.

7. After the loop ends:
Print total students entered.
Print all passed students.
Print all departments available from the tuple using a loop.
Print the student with highest marks.
Use proper formatted output statements.

```python

university_name = "Gift University"
total_students = 1000
scholarship_budget = 50000
passing_marks = 50
students = [
    "Areeha Nasir",
    "Addiya Massora",
    "Saba Junaid",
    "Saba Tariq",
    "Rubab Shafiq",
    "Salwa Yousaf",
    "Uswa Yousaf",
    "Tooba Sadia"
]

# Departments Tuple
departments = ("CS", "SE", "AI", "DS", "IT")

# Variables

passed_students = []

student_count = 0

highest_marks = 0

top_student = ""

# Input Loop

while True:

    student_name = input("Enter student name (or 'exit' to stop): ")
    if student_name.lower() == "exit":
        break
    department = input("Enter department: ")
    marks_obtained = float(input("Enter marks obtained: "))
    attendance_percentage = float(input("Enter attendance percentage: "))
    student_count += 1
    # Department Validation
    if department not in departments:
        print("Invalid Department")
    else:
        # Nested If Conditions
    
        if marks_obtained >= passing_marks and attendance_percentage > 75:
            print("Eligible for Exam")
            passed_students.append(student_name)
            # Scholarship Condition
            if marks_obtained > 85 and attendance_percentage > 80:
                print("Scholarship Candidate")
        elif attendance_percentage < 50 or marks_obtained < 50:
            print("Student Not Eligible")
        else:
            print("Average Performance")
    # Highest Marks Check
    if marks_obtained > highest_marks:
        highest_marks = marks_obtained
        top_student = student_name
# Final Output
print("Final Report ")
print("University Name:", university_name)
print("Total Students Entered:", student_count)
print("Passed Students:", passed_students)
print("Available Departments:")
for dept in departments:
    print(dept)
print("Top Student:", top_student)
print("Highest Marks:", highest_marks)
```

Question 1 — Hospital Management System
Problem Statement

Create a Hospital Management System using Python.

Store the following information in variables:

Hospital name
Total doctors
Emergency budget

Create a tuple for hospital departments:

Cardiology
Neurology
Orthopedic
Emergency

Take the following input from the user:

Patient name
Department
Age
Fever temperature

Use a while loop to continuously take patient data until the user enters "exit".

Conditions
1.

If the entered department is not available in the departments tuple:

Invalid Department
2.

If:

fever temperature > 100
AND
age > 60

Print:

Critical Patient
3.

If:

fever temperature > 100
AND
age <= 60

Print:

Need Observation
4.

Otherwise print:

Normal Patient
Additional Tasks
Store critical patients in a list
Count total patients
Find the patient with the highest fever
Print a final report at the end

```python

hospital_name = "City Hospital"
total_doctors = 20
emergency_budget = 500000

departments = ("Cardiology", "Neurology", "Orthopedic", "Emergency")

# List to store critical patients
critical_patients = []
# Counter for total patients
total_patients = 0
# Highest fever tracking
highest_fever = 0
highest_fever_patient = ""
print("Hospital Management System")
# Loop
while True:
    patient_name = input("Enter Patient Name: ")
    if patient_name.lower() == "exit":
        break
    department = input("Enter Department: ")
    age = int(input("Enter Age: "))
    fever = float(input("Enter Fever Temperature: "))
    # Count patient
    total_patients += 1
    # Department validation
    if department not in departments:
        print("Invalid Department")
    else:
        # Critical
        if fever > 100 and age > 60:
            print("Critical Patient")
            critical_patients.append(patient_name)
        # Observation
        elif fever > 100 and age <= 60:
            print("Need Observation")
        # Normal
        else:
            print("Normal Patient")
    # Highest fever check
    if fever > highest_fever:
        highest_fever = fever
        highest_fever_patient = patient_name
# Final Report
print("Final Report")
print("Hospital:", hospital_name)
print("Total Doctors:", total_doctors)
print("Emergency Budget:", emergency_budget)
print("Total Patients:", total_patients)
print("Critical Patients:", critical_patients)
if total_patients > 0:
    print("Highest Fever Patient:", highest_fever_patient)
    print("Temperature:", highest_fever)
else:
    print("No patient data entered")
```
